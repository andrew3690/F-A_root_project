```python
stocks = pd.read_excel(r"../RAW_DATA/tst/AtivosBR.xlsx",
                       sheet_name="historical_data")
```

The next cells format the stocks excel file

```python
# Renaming column Unnamed to Ticker and set it as index 
stocks.rename(columns = {'symbol':'Ticker'},inplace = True)
stocks.set_index('Ticker')

stocks.columns
```

```python
# #Selecting columns that should be dropped
selected_columns = ['regularMarketSource','maxAge','Unnamed: 0', 
                    'exchange', 'fromCurrency', 'quoteType', 
                    'priceHint', 'fromCurrency', 'toCurrency', 
                    'lastMarket','underlyingSymbol','currency',
                    'quoteSourceName', 'currencySymbol', 'fromCurrency', 'toCurrency']

stocks.drop(columns=selected_columns,inplace=True)
```

```python
stocks
```

```python
string_to_drop = "Quote not found for ticker symbol: "
stocks = stocks[~stocks['Ticker'].str.contains(string_to_drop)]
```

```python
stocks['Ticker'] = stocks['Ticker'].str.replace('.SA', '')
```

```python
stocks
```

```python
stocks.columns
```

```python
stocks.rename(
            columns= {
                'regularMarketChangePercent' : 'Porcentagem de Mudança do Mercado Regular',
                'regularMarketChange' : 'Mudança do Mercado Regular',
                'regularMarketTime' : 'Horário do Mercado Regular',
                'regularMarketPrice' : 'Preço do Mercado Regular',
                'regularMarketDayHigh' : 'Máxima do Mercado Regular',
                'regularMarketDayLow' : 'Mínima do Mercado Regular',
                'regularMarketVolume' : 'Volume do Mercado Regular',
                'regularMarketPreviousClose' : 'Fechamento Anterior do Mercado Regular',
                'regularMarketOpen' : 'Abertura do Mercado Regular',
                'exchangeName' : 'Nome da Bolsa',
                'exchangeDataDelayedBy' : 'Dados da Bolsa Atrasados em',
                'marketState' : 'Estado do Mercado',
                'Ticker' : 'Ticker',
                'shortName' : 'Nome Curto',
                'longName' : 'Nome Longo',
                'marketCap' : 'Capitalização de Mercado'
                }, inplace = True)
```

```python
stocks
```

```python
# stocks.to_parquet(path=r'../RAW_DATA/Parquet/tst/AtivosBr.parquet')

# Define the output directory and filename
output_directory = "../Final/Parquet/tst/"

# Create the output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

output_directory = output_directory + "" + "/AtivosBr.parquet"

stocks.to_parquet(output_directory)
```
